## 1ï¸âƒ£ Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print("âœ… All libraries imported successfully!")
```

---
## 2ï¸âƒ£ Load & Explore the Dataset

```python
ds = pd.read_csv('/content/sample_data/Experience-Salary.csv')

print("ðŸ“„ Dataset Shape:", ds.shape)
print("ðŸ“‹ Column Names:", ds.columns.tolist())
ds.head()
```

```python
print("ðŸ“Š Descriptive Statistics:")
ds.describe().round(2)
```

```python
print("ðŸ” Missing Values:")
print(ds.isnull().sum())
print("\nâœ… No missing values found!" if ds.isnull().sum().sum() == 0 else "âš ï¸ Missing values detected!")
```

---
## 3ï¸âƒ£ Exploratory Data Analysis (EDA)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Exploratory Data Analysis', fontsize=16, fontweight='bold')

# 1. Scatter: Experience vs Salary
axes[0].scatter(ds.iloc[:, 0], ds.iloc[:, 1],
 alpha=0.4, color='steelblue', edgecolors='white', linewidth=0.3, s=15)
axes[0].set_title('Experience vs Salary')
axes[0].set_xlabel('Experience (months)')
axes[0].set_ylabel('Salary (thousands)')
axes[0].grid(True, linestyle='--', alpha=0.4)

# 2. Distribution of Experience
axes[1].hist(ds.iloc[:, 0], bins=30, color='coral', edgecolor='white', linewidth=0.5)
axes[1].set_title('Distribution of Experience')
axes[1].set_xlabel('Experience (months)')
axes[1].set_ylabel('Frequency')
axes[1].grid(True, linestyle='--', alpha=0.4)

# 3. Distribution of Salary
axes[2].hist(ds.iloc[:, 1], bins=30, color='mediumseagreen', edgecolor='white', linewidth=0.5)
axes[2].set_title('Distribution of Salary')
axes[2].set_xlabel('Salary (thousands)')
axes[2].set_ylabel('Frequency')
axes[2].grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()

corr = ds.iloc[:, 0].corr(ds.iloc[:, 1])
print(f"\nðŸ“ˆ Pearson Correlation between Experience and Salary: {corr:.4f}")
```

---
## 4ï¸âƒ£ Data Preprocessing

```python
# Define features (X) and target (y)
x = ds.iloc[:, 0] # Independent: Experience (months)
y = ds.iloc[:, 1] # Dependent: Salary (thousands)

print("âœ… Feature (X) â€” Experience (months):")
print(x.head())
print("\nâœ… Target (y) â€” Salary (thousands):")
print(y.head())
```

```python
# Train-Test Split (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(
 x, y, test_size=0.2, random_state=0
)

print(f"ðŸ“¦ Total samples : {len(x)}")
print(f"ðŸ‹ï¸ Training samples : {len(x_train)} ({len(x_train)/len(x)*100:.0f}%)")
print(f"ðŸ§ª Test samples : {len(x_test)} ({len(x_test)/len(x)*100:.0f}%)")
```

---
## 5ï¸âƒ£ Model Building

### 5.1 Linear Regression

```python
# Train Linear Regression â€” fit BEFORE predict
lr = LinearRegression()
lr.fit(x_train.values.reshape(-1, 1), y_train)

print("âœ… Linear Regression model trained!")
print(f" Intercept (Î²â‚€) : {lr.intercept_:.4f}")
print(f" Coefficient (Î²â‚): {lr.coef_[0]:.4f}")
print(f" Equation: Salary = {lr.intercept_:.2f} + {lr.coef_[0]:.2f} Ã— Experience")
```

```python
# Cross-validation (5-fold)
cv_mse = cross_val_score(
 lr, x.values.reshape(-1, 1), y,
 scoring='neg_mean_squared_error', cv=5
)
cv_r2 = cross_val_score(
 lr, x.values.reshape(-1, 1), y,
 scoring='r2', cv=5
)

print("ðŸ“Š 5-Fold Cross-Validation Results (Linear Regression):")
print(f" Mean MSE : {np.mean(np.abs(cv_mse)):.4f} (Â±{np.std(cv_mse):.4f})")
print(f" Mean RÂ² : {np.mean(cv_r2):.4f} (Â±{np.std(cv_r2):.4f})")
```

### 5.2 Ridge Regression (L2 Regularization)

```python
# Ridge Regression with GridSearchCV
ridge = Ridge()
ridge_params = {'alpha': [1e-15, 1e-10, 1e-3, 1e-2, 5, 8, 10, 111, 100, 120, 500]}

ridge_regressor = GridSearchCV(
 ridge, ridge_params,
 scoring='neg_mean_squared_error', cv=5
)
ridge_regressor.fit(x.values.reshape(-1, 1), y)

print("âœ… Ridge Regression tuned!")
print(f" Best Alpha : {ridge_regressor.best_params_['alpha']}")
print(f" Best MSE : {abs(ridge_regressor.best_score_):.4f}")
```

### 5.3 Lasso Regression (L1 Regularization)

```python
# Lasso Regression with GridSearchCV
lasso = Lasso()
lasso_params = {'alpha': [1e-15, 1e-10, 1e-3, 1e-2, 2, 5, 10, 20, 30, 35, 40, 100, 200]}

lasso_regressor = GridSearchCV(
 lasso, lasso_params,
 scoring='neg_mean_squared_error', cv=5
)
lasso_regressor.fit(x.values.reshape(-1, 1), y)

print("âœ… Lasso Regression tuned!")
print(f" Best Alpha : {lasso_regressor.best_params_['alpha']}")
print(f" Best MSE : {abs(lasso_regressor.best_score_):.4f}")
```

---
## 6ï¸âƒ£ Model Evaluation

```python
x_test_r = x_test.values.reshape(-1, 1)
x_train_r = x_train.values.reshape(-1, 1)

y_pred_train = lr.predict(x_train_r)
y_pred_test = lr.predict(x_test_r)

train_mse = mean_squared_error(y_train, y_pred_train)
test_mse = mean_squared_error(y_test, y_pred_test)
train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)
train_mae = mean_absolute_error(y_train, y_pred_train)
test_mae = mean_absolute_error(y_test, y_pred_test)
train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

print("="*55)
print(f"{'Metric':15} {'Test':>15}")
print("="*55)
print(f"{'MSE':15.4f} {test_mse:>15.4f}")
print(f"{'RMSE':15.4f} {test_rmse:>15.4f}")
print(f"{'MAE':15.4f} {test_mae:>15.4f}")
print(f"{'RÂ² Score':15.4f} {test_r2:>15.4f}")
print("="*55)
print(f"\nðŸ’¡ The model explains {test_r2*100:.1f}% of salary variance on unseen data.")
```

```python
ridge_best = Ridge(alpha=ridge_regressor.best_params_['alpha'])
ridge_best.fit(x_train_r, y_train)
ridge_r2 = r2_score(y_test, ridge_best.predict(x_test_r))
ridge_mse = mean_squared_error(y_test, ridge_best.predict(x_test_r))

lasso_best = Lasso(alpha=lasso_regressor.best_params_['alpha'])
lasso_best.fit(x_train_r, y_train)
lasso_r2 = r2_score(y_test, lasso_best.predict(x_test_r))
lasso_mse = mean_squared_error(y_test, lasso_best.predict(x_test_r))

print("ðŸ“Š Model Comparison (Test Set):")
print("="*50)
print(f"{'Model':10} {'RÂ² Score':>12}")
print("="*50)
print(f"{'Linear Regression':10.4f} {test_r2:>12.4f}")
print(f"{'Ridge Regression':10.4f} {ridge_r2:>12.4f}")
print(f"{'Lasso Regression':10.4f} {lasso_r2:>12.4f}")
print("="*50)
print("\nðŸ’¡ All three models perform identically â€” regularization adds no benefit")
print(" for this simple univariate, low-noise dataset.")
```

### 7.3 Residual Analysis

```python
residuals = y_test.values - y_pred_test

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle('Residual Analysis', fontsize=14, fontweight='bold')

# Residuals vs Fitted
axes[0].scatter(y_pred_test, residuals,
 color='steelblue', alpha=0.55, edgecolors='navy',
 linewidth=0.4, s=18)
axes[0].axhline(y=0, color='red', linestyle='--', linewidth=2)
axes[0].set_title('Residuals vs Fitted Values')
axes[0].set_xlabel('Fitted Values (Predicted Salary)')
axes[0].set_ylabel('Residuals')
axes[0].grid(True, linestyle='--', alpha=0.4)

# Residual distribution
axes[1].hist(residuals, bins=30,
 color='mediumseagreen', edgecolor='white', linewidth=0.5)
axes[1].axvline(x=0, color='red', linestyle='--', linewidth=2)
axes[1].set_title('Distribution of Residuals')
axes[1].set_xlabel('Residuals')
axes[1].set_ylabel('Frequency')
axes[1].grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()

print(f"Mean of Residuals : {np.mean(residuals):.6f} (should be â‰ˆ 0 â€” confirms unbiased fit)")
print(f"Std of Residuals : {np.std(residuals):.4f}")
```
