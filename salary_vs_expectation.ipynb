## 1️⃣ Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler

print("✅ All libraries imported successfully!")
```

---
## 2️⃣ Load & Explore the Dataset

```python
# Load dataset
ds = pd.read_csv('/content/sample_data/Experience-Salary.csv')

print("📄 Dataset Shape:", ds.shape)
print("\n📋 Column Names:", ds.columns.tolist())
ds.head(10)
```

```python
# Basic statistics
print("📊 Descriptive Statistics:")
ds.describe().round(2)
```

```python
# Check for missing values
print("🔍 Missing Values:")
print(ds.isnull().sum())
print("\n✅ No missing values found!" if ds.isnull().sum().sum() == 0 else "⚠️ Missing values detected!")
```

---
## 3️⃣ Exploratory Data Analysis (EDA)

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Exploratory Data Analysis', fontsize=16, fontweight='bold')

# 1. Scatter plot: Experience vs Salary
axes[0].scatter(ds.iloc[:, 0], ds.iloc[:, 1], alpha=0.5, color='steelblue', edgecolors='white', linewidth=0.3)
axes[0].set_title('Experience vs Salary')
axes[0].set_xlabel('Experience (months)')
axes[0].set_ylabel('Salary (thousands)')
axes[0].grid(True, linestyle='--', alpha=0.5)

# 2. Distribution of Experience
axes[1].hist(ds.iloc[:, 0], bins=30, color='coral', edgecolor='white', linewidth=0.5)
axes[1].set_title('Distribution of Experience')
axes[1].set_xlabel('Experience (months)')
axes[1].set_ylabel('Frequency')
axes[1].grid(True, linestyle='--', alpha=0.5)

# 3. Distribution of Salary
axes[2].hist(ds.iloc[:, 1], bins=30, color='mediumseagreen', edgecolor='white', linewidth=0.5)
axes[2].set_title('Distribution of Salary')
axes[2].set_xlabel('Salary (thousands)')
axes[2].set_ylabel('Frequency')
axes[2].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

# Correlation
corr = ds.iloc[:, 0].corr(ds.iloc[:, 1])
print(f"\n📈 Pearson Correlation between Experience and Salary: {corr:.4f}")
```

---
## 4️⃣ Data Preprocessing

```python
# Define features (X) and target (y)
x = ds.iloc[:, 0]   # Independent: Experience (months)
y = ds.iloc[:, 1]   # Dependent:   Salary (thousands)

print("✅ Feature (X) — Experience (months):")
print(x.head())
print("\n✅ Target (y) — Salary (thousands):")
print(y.head())
```

```python
# Train-Test Split (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=0
)

print(f"📦 Total samples    : {len(x)}")
print(f"🏋️  Training samples : {len(x_train)} ({len(x_train)/len(x)*100:.0f}%)")
print(f"🧪 Test samples     : {len(x_test)} ({len(x_test)/len(x)*100:.0f}%)")
```

---
## 5️⃣ Model Building

### 5.1 Linear Regression

```python
# Train Linear Regression
lr = LinearRegression()
lr.fit(x_train.values.reshape(-1, 1), y_train)

print("✅ Linear Regression model trained!")
print(f"   Intercept (β₀) : {lr.intercept_:.4f}")
print(f"   Coefficient (β₁): {lr.coef_[0]:.4f}")
print(f"   Equation: Salary = {lr.intercept_:.2f} + {lr.coef_[0]:.2f} × Experience")
```

```python
# Cross-validation score (5-fold)
cv_mse = cross_val_score(
    lr, x.values.reshape(-1, 1), y,
    scoring='neg_mean_squared_error', cv=5
)
cv_r2 = cross_val_score(
    lr, x.values.reshape(-1, 1), y,
    scoring='r2', cv=5
)

print("📊 5-Fold Cross-Validation Results (Linear Regression):")
print(f"   Mean MSE : {np.mean(np.abs(cv_mse)):.4f}  (±{np.std(cv_mse):.4f})")
print(f"   Mean R²  : {np.mean(cv_r2):.4f}  (±{np.std(cv_r2):.4f})")
```

### 5.2 Ridge Regression (L2 Regularization)

```python
# Ridge Regression with GridSearchCV
ridge = Ridge()
ridge_params = {'alpha': [1e-15, 1e-10, 1e-3, 1e-2, 5, 8, 10, 111, 100, 120, 500]}

ridge_regressor = GridSearchCV(
    ridge, ridge_params,
    scoring='neg_mean_squared_error', cv=5
)
ridge_regressor.fit(x.values.reshape(-1, 1), y)

print("✅ Ridge Regression tuned!")
print(f"   Best Alpha : {ridge_regressor.best_params_['alpha']}")
print(f"   Best MSE   : {abs(ridge_regressor.best_score_):.4f}")
```

### 5.3 Lasso Regression (L1 Regularization)

```python
# Lasso Regression with GridSearchCV
lasso = Lasso()
lasso_params = {'alpha': [1e-15, 1e-10, 1e-3, 1e-2, 2, 5, 10, 20, 30, 35, 40, 100, 200]}

lasso_regressor = GridSearchCV(
    lasso, lasso_params,
    scoring='neg_mean_squared_error', cv=5
)
lasso_regressor.fit(x.values.reshape(-1, 1), y)

print("✅ Lasso Regression tuned!")
print(f"   Best Alpha : {lasso_regressor.best_params_['alpha']}")
print(f"   Best MSE   : {abs(lasso_regressor.best_score_):.4f}")
```

---
## 6️⃣ Model Evaluation

```python
# Generate predictions on test set
x_test_reshaped = x_test.values.reshape(-1, 1)
x_train_reshaped = x_train.values.reshape(-1, 1)

y_pred_train = lr.predict(x_train_reshaped)
y_pred_test  = lr.predict(x_test_reshaped)

# Metrics
train_mse  = mean_squared_error(y_train, y_pred_train)
test_mse   = mean_squared_error(y_test,  y_pred_test)
train_rmse = np.sqrt(train_mse)
test_rmse  = np.sqrt(test_mse)
train_mae  = mean_absolute_error(y_train, y_pred_train)
test_mae   = mean_absolute_error(y_test,  y_pred_test)
train_r2   = r2_score(y_train, y_pred_train)
test_r2    = r2_score(y_test,  y_pred_test)

print("="*55)
print(f"{'Metric':<20} {'Train':>15} {'Test':>15}")
print("="*55)
print(f"{'MSE':<20} {train_mse:>15.4f} {test_mse:>15.4f}")
print(f"{'RMSE':<20} {train_rmse:>15.4f} {test_rmse:>15.4f}")
print(f"{'MAE':<20} {train_mae:>15.4f} {test_mae:>15.4f}")
print(f"{'R² Score':<20} {train_r2:>15.4f} {test_r2:>15.4f}")
print("="*55)

print(f"\n💡 Interpretation: The model explains {test_r2*100:.1f}% of the variance in salary on unseen data.")
```

```python
# Model comparison table
ridge_best = Ridge(alpha=ridge_regressor.best_params_['alpha'])
ridge_best.fit(x_train_reshaped, y_train)
ridge_r2 = r2_score(y_test, ridge_best.predict(x_test_reshaped))
ridge_mse = mean_squared_error(y_test, ridge_best.predict(x_test_reshaped))

lasso_best = Lasso(alpha=lasso_regressor.best_params_['alpha'])
lasso_best.fit(x_train_reshaped, y_train)
lasso_r2 = r2_score(y_test, lasso_best.predict(x_test_reshaped))
lasso_mse = mean_squared_error(y_test, lasso_best.predict(x_test_reshaped))

print("📊 Model Comparison (Test Set):")
print("="*50)
print(f"{'Model':<20} {'MSE':>12} {'R² Score':>12}")
print("="*50)
print(f"{'Linear Regression':<20} {test_mse:>12.4f} {test_r2:>12.4f}")
print(f"{'Ridge Regression':<20} {ridge_mse:>12.4f} {ridge_r2:>12.4f}")
print(f"{'Lasso Regression':<20} {lasso_mse:>12.4f} {lasso_r2:>12.4f}")
print("="*50)
```

---
## 7️⃣ Visualization

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Linear Regression — Training vs Test Set', fontsize=15, fontweight='bold')

# --- Training Set ---
axes[0].scatter(x_train, y_train, color='gold', edgecolors='darkorange',
                linewidth=0.5, alpha=0.7, label='Training data', zorder=2)
x_line = np.linspace(x.min(), x.max(), 200).reshape(-1, 1)
axes[0].plot(x_line, lr.predict(x_line), color='royalblue',
             linewidth=2.5, label='Regression line')
axes[0].set_title('Training Set')
axes[0].set_xlabel('Experience (months)')
axes[0].set_ylabel('Salary (thousands)')
axes[0].legend()
axes[0].grid(True, linestyle='--', alpha=0.4)

# --- Test Set ---
axes[1].scatter(x_test, y_test, color='tomato', edgecolors='darkred',
                linewidth=0.5, alpha=0.7, label='Test data', zorder=2)
axes[1].plot(x_line, lr.predict(x_line), color='royalblue',
             linewidth=2.5, label='Regression line')
axes[1].set_title('Test Set')
axes[1].set_xlabel('Experience (months)')
axes[1].set_ylabel('Salary (thousands)')
axes[1].legend()
axes[1].grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
# Actual vs Predicted plot
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(y_test, y_pred_test, color='mediumpurple', edgecolors='indigo',
           alpha=0.6, linewidth=0.5, label='Predicted vs Actual')
ax.plot([y.min(), y.max()], [y.min(), y.max()],
        'r--', linewidth=2, label='Perfect prediction line')
ax.set_title('Actual vs Predicted Salary (Test Set)', fontsize=13, fontweight='bold')
ax.set_xlabel('Actual Salary (thousands)')
ax.set_ylabel('Predicted Salary (thousands)')
ax.legend()
ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Residuals plot
residuals = y_test.values - y_pred_test

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle('Residual Analysis', fontsize=14, fontweight='bold')

# Residuals vs Fitted
axes[0].scatter(y_pred_test, residuals, color='steelblue', alpha=0.6, edgecolors='navy', linewidth=0.4)
axes[0].axhline(y=0, color='red', linestyle='--', linewidth=2)
axes[0].set_title('Residuals vs Fitted Values')
axes[0].set_xlabel('Fitted Values (Predicted Salary)')
axes[0].set_ylabel('Residuals')
axes[0].grid(True, linestyle='--', alpha=0.4)

# Residuals distribution
axes[1].hist(residuals, bins=30, color='mediumseagreen', edgecolor='white', linewidth=0.5)
axes[1].axvline(x=0, color='red', linestyle='--', linewidth=2)
axes[1].set_title('Distribution of Residuals')
axes[1].set_xlabel('Residuals')
axes[1].set_ylabel('Frequency')
axes[1].grid(True, linestyle='--', alpha=0.4)

plt.tight_layout()
plt.show()

print(f"Mean of Residuals : {np.mean(residuals):.6f}  (should be ≈ 0)")
print(f"Std of Residuals  : {np.std(residuals):.4f}")
```

## 8️⃣ Make New Predictions

```python
# Predict salary for new experience values
new_experience = np.array([12, 24, 36, 48, 60]).reshape(-1, 1)  # months
predicted_salaries = lr.predict(new_experience)

print("🔮 Salary Predictions for New Experience Values:")
print("="*45)
print(f"{'Experience (months)':<25} {'Predicted Salary (K)':>18}")
print("="*45)
for exp, sal in zip(new_experience.flatten(), predicted_salaries):
    print(f"{exp:<25} {sal:>18.2f}")
print("="*45)
```
